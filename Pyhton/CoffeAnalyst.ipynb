## Item

Langkah pertama pada feature item dilakukan pengecekan data seperti data unique dan total missing values

```python
unikItem = Data1['Item'].unique()
misvalItem = Data1['Item'].isna().sum()

print('Item Feauture')
print(f'Data Unik: \n{unikItem}\n')
print(f'Total Missing Value: \n{misvalItem}')
```

kemudian membuat sebuah variabel untuk 
1. mengisi data null dengan 'UNKNOWN' pada kolom 'Item'
2. menggabungkan baris yang berisi 'UNKNOWN' dan 'ERROR' pada kolom 'Item' menjadi 'Unknown' agar mudah untuk di analisis selanjutnya

```python
misvalItem1 = Data1['Item'] = Data1['Item'].fillna('UNKNOWN')
Data1['Item'] = Data1['Item'].apply(
    lambda x:'Unknown' if str(x) in ['UNKNOWN','ERROR']
    else str(x)
)

print('Isi data null dengan UNKNOWN dan gabungkan data UNKNOWN dan ERROR menjadi Unknown\n')
print('Data setelah di diisi dan digabungkan')
Data1['Item'].unique()
```

cek apakah masih ada data null

```python
Data1['Item'].isna().sum()
```

variabel 'total_Item' digunakan untuk menghitung jumlah data pada masing-masing Itemnya

```python
total_Item = Data1['Item'].value_counts()
print(f'Total setiap Itemnya{total_Item}')
```

**Menambahkan feature baru(Type, Category, Cost Per Item, Total Cost, Profit) untuk menambah insight**

## Type

kelompokkan setiap Item menjadi beberapa Type seperti Main Dish, Dessert, Cold Baverage, dan Hot Baverage

```python
Jenis_Hidangan = {
    'Salad': 'Main Dish',
    'Sandwich': 'Main Dish',
    'Cake': 'Dessert',
    'Cookie': 'Dessert',
    'Smoothie': 'Cold Beverage',
    'Juice': 'Cold Beverage',
    'Coffee': 'Hot Beverage',
    'Tea': 'Hot Beverage',
    'Unknown': 'Unknown'
}

Data1['Type'] = Data1['Item'].map(Jenis_Hidangan) 
```

cek total data unik pada feature 'Type'

```python
total_Jenis = Data1['Type'].value_counts(dropna=False)
print(total_Jenis)
```

## Category

bagi setiap Item menjadi 2 kategori yaitu Food dan Drink

```python
Kategori_Hidangan = {
    'Salad': 'Food',
    'Sandwich': 'Food',
    'Cake': 'Food',
    'Cookie': 'Food',
    'Smoothie': 'Drink',
    'Juice': 'Drink',
    'Coffee': 'Drink',
    'Tea': 'Drink',
    'Unknown': 'Unknown'
}

Data1['Category'] = Data1['Item'].map(Kategori_Hidangan) 
```

cek total data unik pada feature 'Category'

```python
total_Kategori = Data1['Category'].value_counts(dropna=False)
print(total_Kategori)
```

## Cost Per Item

tambahkan harga bahan baku pembuatan pada setiap Itemnya untuk menambahkan insight baru

```python
cost_map = {
    'Salad': 2.0,
    'Sandwich': 1.6,
    'Cake': 1.05,
    'Cookie': 0.3,
    'Juice': 1.05,
    'Coffee': 0.5,
    'Smoothie': 1.6,
    'Tea': 0.3,
    'Unknown': 0
}

Data1['Cost Per Item'] = Data1['Item'].map(cost_map)
Data1['Cost Per Item'] = Data1['Cost Per Item'].astype(float)
Data1['Cost Per Item'].unique()
```

## Total Cost

buat perhitungan harga total bahan baku dari Item yang dibeli pada setiap transaksinya, dengan mengkalikan 'Quantity' dan 'Cost Per Item'

```python
Data1['Total Cost'] = Data1['Cost Per Item'] * Data1['Quantity']
Data1['Total Cost'] = Data1['Total Cost'].round(2)
Data1['Total Cost'].unique()
```

## Profit

feature baru 'Profit' digunakan untuk mengetahui keuntungan bersih dari masing-masing Item yang dijual

```python
Data1['Profit'] = Data1['Total Spent'] - Data1['Total Cost']
Data1['Profit'].unique()
```

```python
Data1.info()
```

dari hasil data unik di atas terdapat bebarapa data yang salah perhitungan, maka dilakukan penghapusan data yang tidak sesusai

```python
Data1 = Data1[Data1['Profit'] >= 0]
Data1['Profit'].unique()
```
