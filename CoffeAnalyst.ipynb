## Cost Per Unit

```python
cost_map = {
    'Salad': 2.0,
    'Sandwich': 1.6,
    'Cake': 1.05,
    'Cookie': 0.3,
    'Juice': 1.05,
    'Coffee': 0.5,
    'Smoothie': 1.6,
    'Tea': 0.3,
    'Unknown': 0
}

Data1['Cost Per Item'] = Data1['Item'].map(cost_map)
Data1['Cost Per Item'] = Data1['Cost Per Item'].astype(float)
Data1['Cost Per Item'].unique()
```

## Total Cost

```python
Data1['Total Cost'] = Data1['Cost Per Item'] * Data1['Quantity']
Data1['Total Cost'] = Data1['Total Cost'].astype(float)
Data1['Total Cost'].unique()
```

## Profit

```python
Data1['Profit'] = Data1['Total Spent'] - Data1['Total Cost']
Data1['Profit'].unique()
```
